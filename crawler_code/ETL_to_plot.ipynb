```python
import pandas, os
data_base_path = os.path.join(os.path.dirname(os.getcwd()), "data_base", "data_base.csv")
df = pandas.read_csv(data_base_path)
df
```

# Pandas map solve

```python
import numpy
def k_unit_translate(var) : 
    try : 
        if "K" in var : 
            var = float(var.split("K")[0])*1000
            return var
        else : 
            return var
    except : 
        return var

def text_in_var(var) : 
    try : 
        var = float(var)
        return var
    except : 
        var = numpy.nan
        return var
        
var_list = ["like-count", "comment_count", "collection", "share_count"]
for var in var_list : 
    df[var] = df[var].map(k_unit_translate)
    df[var] = df[var].map(text_in_var)
```

# Parser_time solve

```python
# 爬取時間檢查
df['parser_time'] = pandas.to_datetime(df['parser_time'])
print(f"爬取開始時間為 : {df['parser_time'].min()}")
print(f"爬取終止時間為 : {df['parser_time'].max()}")
```

```python
from matplotlib.font_manager import FontProperties
import matplotlib.pyplot as plt
import numpy
import warnings
import pandas
import traceback
import os

# 設置中文字體路徑
font_path = 'ttf\SimHei.ttf'
myfont = FontProperties(fname=r'ttf\chinese.simhei.ttf')

# 關閉因 emoji 無法正常顯示在標題的警告
warnings.filterwarnings("ignore", message="Glyph .* missing from current font")

# set plot-var-list
plot_var_list = ["like-count", "comment_count", "collection", "share_count"]

error_info_list = []
for post_id in df["post_id"].unique() : 
    for plot_var in plot_var_list : 
        try :
            post_df = df[df["post_id"] == post_id].copy() 
            if len(post_df[post_df["post_title"].notna()]["post_title"].values) == 0 : # title field may not necessarily have a value.
                continue
            
            post_df[plot_var] = post_df[plot_var].astype(float)
            plot_title = post_df[post_df["post_title"].notna()]["post_title"].values[0]
            values_axis = post_df[plot_var]
            time_axis = post_df["parser_time"]

            # 繪製折線圖
            plt.plot(time_axis, values_axis, label=plot_var)
            plt.title(plot_title.split("#")[0], fontproperties=myfont)
            plt.xlabel('parser_time', fontproperties=myfont)

            # 額外設定
            time_axis_ticks = pandas.date_range(start=post_df["parser_time"].min(), end = post_df["parser_time"].max(), freq='h')
            time_axis_ticks = time_axis_ticks[::12]
            plt.xticks(time_axis_ticks, [tick.strftime('%m-%d %H') for tick in time_axis_ticks], rotation=45)
            plt.legend()

            # save line-plot
            save_path = os.path.join(os.path.dirname(os.getcwd()), "line_plot", f"{post_id}")
            if not os.path.exists(save_path):
                os.makedirs(save_path)
            
            plt.savefig(os.path.join(save_path, f'{plot_var}.png'), dpi=300)
            plt.close()
        except : 
            error_info = {"post_id" : post_id, "plot_var" : plot_var, "traceback" : traceback.format_exc()}
            error_info_list.append(error_info)
            continue

error_info_df = pandas.DataFrame(error_info_list)
error_info_df.to_json("plot_error.json")
```
